```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data_2019 = pd.read_csv('flujo-vehicular-2019.csv', sep =',')
data_2018 = pd.read_csv('flujo-vehicular-2018.csv', sep =',')
data_2017 = pd.read_csv('flujo-vehicular-2017.csv', sep =',')
```

```python
plt.figure(figsize=(20,8))
    
for estacion in data_2019.estacion.unique():
    
    ds = data_2019[(data_2019.forma_pago == 'EFECTIVO') & (data_2019.estacion == f'{estacion}')].copy()

    ds['date'] = pd.to_datetime(ds.fecha) + pd.to_timedelta(ds.hora_fin, unit = 'h')

    ds = ds.resample('D', on = 'date').sum()

    plt.plot(ds.index, ds.cantidad_pasos, label = f'{estacion}', lw=0.8); plt.legend()
```
